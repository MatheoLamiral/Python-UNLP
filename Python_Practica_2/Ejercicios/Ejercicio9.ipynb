### Ejercicio 9 

Escriba un programa que solicite por teclado una palabra y calcule el valor de la misma
dada la siguiente tabla de valores del juego Scrabble:

| Letra                         | Valor         |
| ----------------------------- | ------------- |
| A, E, I, O, U, L, N, R, S, T  | 1             |
| D,G                           |  2            |
| B,C,M,P                       | 3             |
| F,H,V,W,Y                     | 4             |
| K                             | 5             |
| J,X                           | 8             |
| Q,Z                           | 10            |

*Tenga en cuenta qu√© estructura elige para guardar estos valores en Python

__Ejemplo 1__

- Palabra: "solo"
- valor: 4

__Ejemplo 2__

- Palabra: "tomate"
- valor: 8


```python
values = {"A,E,I,O,U,L,N,R,S,T":1,
           "D,G":2,
           "B,C,M,P":3,
           "F,H,V,W,Y":4,
           "K":5,
           "J,X":8,
           "Q,Z":10}

word = input("Ingrese una palabra:").upper()

word_points = 0
for char in word:
    for key in values:
        if char in key:
            word_points += values[key]
            break
    
print(f"El valor de la palabra es {word_points}")
```
